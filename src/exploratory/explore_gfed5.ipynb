```python
import os
import glob
import numpy as np
import xarray as xr

DATA_DIR = os.path.join('data', 'fire', 'gfed5.1', 'GFED5.1_ecosystem')

# Find and open the first .nc file
pattern = os.path.join(DATA_DIR, '*')
all_files = sorted([p for p in glob.glob(pattern) if os.path.isfile(p) and (p.lower().endswith('.nc') or '.nc' in p.lower())])

if all_files:
    first_file = all_files[0]
    print(f"Opening first NetCDF file: {first_file}\n")
    ds = xr.open_dataset(first_file)
    
    # Check if basis_regions exists
    if 'basis_regions' in ds.data_vars or 'basis_regions' in ds.coords:
        print("=" * 80)
        print("BASIS_REGIONS VARIABLE")
        print("=" * 80)
        br = ds['basis_regions']
        print(f"\nShape: {br.shape}")
        print(f"Data type: {br.dtype}")
        print(f"Dimensions: {br.dims}")
        print(f"\nUnique values in basis_regions:")
        unique_vals = np.unique(br.values)
        print(f"  Count: {len(unique_vals)}")
        print(f"  Values: {sorted(unique_vals)}")
        
        print(f"\nValue counts:")
        for val in sorted(unique_vals):
            count = np.sum(br.values == val)
            print(f"  {val}: {count} occurrences")
        
        print(f"\nSample 5×5 slice:")
        print(br.values[:5, :5])
    else:
        print("'basis_regions' not found in dataset.")
        print(f"Available data variables: {list(ds.data_vars.keys())}")
        print(f"Available coordinates: {list(ds.coords.keys())}")
    
    ds.close()
else:
    print(f"No .nc files found in {DATA_DIR}")
```

```python
# Explore GFED5.1 ecosystem NetCDF files
# - Lists files
# - Opens a small sample with xarray
# - Prints variable/dimension/time summaries
# - Saves a summary CSV and quick plots to outputs/

import os
import glob
import pandas as pd
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

DATA_DIR = os.path.join('data', 'fire', 'gfed5.1', 'GFED5.1_ecosystem')
os.makedirs('outputs', exist_ok=True)

# Find NetCDF files (support .nc and .nc4 and compressed variants)
pattern = os.path.join(DATA_DIR, '*')
all_files = sorted([p for p in glob.glob(pattern) if os.path.isfile(p) and (p.lower().endswith('.nc') or '.nc' in p.lower())])
print(f"Found {len(all_files)} candidate files in {DATA_DIR}")
for i, f in enumerate(all_files[:20], 1):
    print(f"{i:2d}. {f}")

# Heuristic helpers
def find_lat_lon(ds):
    lat_names = [n for n in ds.coords.keys() if n.lower() in ('lat','latitude','y')]
    lon_names = [n for n in ds.coords.keys() if n.lower() in ('lon','longitude','x')]
    lat = lat_names[0] if lat_names else None
    lon = lon_names[0] if lon_names else None
    return lat, lon

summary_rows = []
# Limit how many files we open to avoid long runs; examine up to first 6 files
for fp in all_files[:6]:
    try:
        ds = xr.open_dataset(fp)
    except Exception as e:
        print(f"Could not open {fp}: {e}")
        continue

    vars_list = list(ds.data_vars.keys())
    dims = dict(ds.dims)
    lat_name, lon_name = find_lat_lon(ds)

    time_start = None
    time_end = None
    if 'time' in ds.coords:
        try:
            tvals = ds['time'].values
            if len(tvals) > 0:
                time_start = np.min(tvals)
                time_end = np.max(tvals)
        except Exception:
            time_start = None
            time_end = None

    # pick first numeric data variable to summarize
    example_var = vars_list[0] if vars_list else None
    example_mean = None
    try:
        if example_var is not None:
            # try to compute mean over time and space if dims present
            dims_to_reduce = []
            for d in ds[example_var].dims:
                if d.lower() in ('time','lat','latitude','y','lon','longitude','x'):
                    dims_to_reduce.append(d)
            if dims_to_reduce:
                example_mean = float(ds[example_var].mean(dim=dims_to_reduce).values)
            else:
                example_mean = float(ds[example_var].mean().values)
    except Exception:
        example_mean = None

    summary_rows.append({
        'file': os.path.relpath(fp),
        'n_vars': len(vars_list),
        'vars': ','.join(vars_list),
        'dims': str(dims),
        'lat_coord': lat_name,
        'lon_coord': lon_name,
        'time_start': str(time_start),
        'time_end': str(time_end),
        'example_var': example_var,
        'example_mean': example_mean
    })

    ds.close()

if summary_rows:
    summary_df = pd.DataFrame(summary_rows)
    out_csv = os.path.join('outputs', 'gfed5_ecosystem_summary.csv')
    summary_df.to_csv(out_csv, index=False)
    print(f"Wrote dataset summary to {out_csv}")
    print(summary_df)
else:
    print("No summaries created (no readable NetCDF files found or all failed to open).")

# Quick plotting: spatial-mean time series and a sample map from the first readable file
if all_files:
    for fp in all_files:
        try:
            ds = xr.open_dataset(fp)
            break
        except Exception:
            ds = None
    if ds is None:
        print("No file could be opened for plotting.")
    else:
        print(f"Using {fp} for quick plots")
        vars_list = list(ds.data_vars.keys())
        if not vars_list:
            print("No data variables available in dataset for plotting.")
        else:
            var = vars_list[2]
            print(f"Selected variable for quick plots: {var}")

            # Identify lat/lon dims
            lat_name, lon_name = find_lat_lon(ds)
            time_dim = 'time' if 'time' in ds.coords else None

            # 1) Time series of spatial mean
            try:
                if time_dim is not None and lat_name and lon_name:
                    ts = ds[var].mean(dim=[lat_name, lon_name])
                    plt.figure(figsize=(8,3))
                    ts.plot.line(marker='o')
                    plt.title(f"Spatial mean time series: {var}")
                    plt.tight_layout()
                    png = os.path.join('outputs', 'gfed5_example_timeseries.png')
                    plt.savefig(png, dpi=200, bbox_inches='tight')
                    plt.show()
                    print(f"Saved time series plot to {png}")
                else:
                    print("Cannot compute spatial-mean time series (missing time or spatial coords).")
            except Exception as e:
                print(f"Error plotting time series: {e}")

            # 2) Sample map for the first time slice (if spatial dims exist)
            try:
                if lat_name and lon_name and time_dim in ds.coords:
                    first_slice = ds[var].isel({time_dim: 0})
                    plt.figure(figsize=(6,5))
                    im = first_slice.plot(cmap='viridis')
                    plt.title(f"Sample map (first time) for {var}")
                    png2 = os.path.join('outputs', 'exploratory', 'gfed', 'gfed5_example_map_first.png')
                    plt.savefig(png2, dpi=200, bbox_inches='tight')
                    plt.show()
                    print(f"Saved sample map to {png2}")
                elif lat_name and lon_name:
                    # no time dim; plot whole variable
                    plt.figure(figsize=(6,5))
                    im = ds[var].plot(cmap='viridis')
                    plt.title(f"Sample map for {var}")
                    png2 = os.path.join('outputs', 'exploratory', 'gfed', 'gfed5_example_map.png')
                    plt.savefig(png2, dpi=200, bbox_inches='tight')
                    plt.show()
                    print(f"Saved sample map to {png2}")
                else:
                    print("No spatial coordinates found; skipping sample map.")
            except Exception as e:
                print(f"Error plotting sample map: {e}")

            ds.close()
else:
    print("No files to inspect.")

print('\nExploration complete. If packages are missing (xarray, netCDF4), install via:')
print('  conda install -c conda-forge xarray netcdf4 rioxarray dask matplotlib')
print('or')
print('  pip install xarray netCDF4 rioxarray dask matplotlib')
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the annual emissions CSV
csv_path = os.path.join('outputs', 'gfed_emissions_and_ba_in_alaska.csv')
df = pd.read_csv(csv_path)

# Filter to 2002-2023
df = df[(df['year'] >= 2002) & (df['year'] <= 2023)]

# Convert units
df['burned_area_mha'] = df['burned_area'] / 1e10  # m² to megahectares (Mha)
df['carbon_emissions_tg'] = df['carbon_emissions'] / 1e12  # grams to Tg (teragrams)

# Create plots
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# ===== Plot 1: Burned area by year =====
ax = axes[0]
ax.plot(df['year'], df['burned_area_mha'], marker='o', linewidth=2, markersize=6, color='#d62728', label='Annual')

# Calculate trend line and statistics
z1 = np.polyfit(df['year'], df['burned_area_mha'], 1)
p1 = np.poly1d(z1)
trend_line1 = p1(df['year'])
ax.plot(df['year'], trend_line1, '--', linewidth=2, color='#d62728', alpha=0.7, label='Trend')

# Add average line
avg1 = df['burned_area_mha'].mean()
ax.axhline(y=avg1, color='gray', linestyle=':', linewidth=1.5, alpha=0.6, label='Long-term mean')

# Add text with statistics
slope1 = z1[0]
text_str1 = f'Slope: {slope1:.4f} Mha/year\nMean: {avg1*1000000:.0f} ha'
ax.text(0.02, 0.98, text_str1, transform=ax.transAxes, fontsize=10, 
        verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

ax.set_xlabel('Year')
ax.set_ylabel('Burned Area (Mha)')
ax.set_title('Annual Burned Area in Alaska (GFED5.1)')
ax.grid(True, alpha=0.3)
ax.legend(loc='lower right')

# ===== Plot 2: Carbon emissions by year =====
ax = axes[1]
ax.plot(df['year'], df['carbon_emissions_tg'], marker='s', linewidth=2, markersize=6, color='#1f77b4', label='Annual')

# Calculate trend line and statistics
z2 = np.polyfit(df['year'], df['carbon_emissions_tg'], 1)
p2 = np.poly1d(z2)
trend_line2 = p2(df['year'])
ax.plot(df['year'], trend_line2, '--', linewidth=2, color='#1f77b4', alpha=0.7, label='Trend')

# Add average line
avg2 = df['carbon_emissions_tg'].mean()
ax.axhline(y=avg2, color='gray', linestyle=':', linewidth=1.5, alpha=0.6, label='Long-term mean')

# Add text with statistics
slope2 = z2[0]
text_str2 = f'Slope: {slope2:.4f} Tg/year\nMean: {avg2:.4f} Tg'
ax.text(0.02, 0.98, text_str2, transform=ax.transAxes, fontsize=10,
        verticalalignment='top', bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))

ax.set_xlabel('Year')
ax.set_ylabel('Carbon Emissions (Tg)')
ax.set_title('Annual Carbon Emissions in Alaska (GFED5.1)')
ax.grid(True, alpha=0.3)
ax.legend(loc='lower right')

plt.tight_layout()
plt.savefig(os.path.join('outputs', 'exploratory', 'gfed', 'annual_emissions_trends.png'), dpi=300, bbox_inches='tight')

plt.show()
print(f"\nPlot saved to outputs/exploratory/gfed/annual_emissions_trends.png")
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read both CSV files
csv_path1 = os.path.join('outputs', 'gfed_emissions_and_ba_in_alaska.csv')
csv_path2 = os.path.join('outputs', 'gfed_emissions_in_alaska_by_gas_gram.csv')

df1 = pd.read_csv(csv_path1)
df2 = pd.read_csv(csv_path2)

# Filter to 2002-2023
df1 = df1[(df1['year'] >= 2002) & (df1['year'] <= 2023)]
df2 = df2[(df2['year'] >= 2002) & (df2['year'] <= 2023)]

# Convert units
df1['carbon_emissions_tg'] = df1['carbon_emissions'] / 1e12  # grams to Tg
df2['carbon_c_tg'] = df2['C'] / 1e12  # grams to Tg

# Create plot comparing the two carbon emission series
fig, ax = plt.subplots(figsize=(11, 6))

# Plot 1: Carbon emissions (ecosystem summary)
ax.plot(df1['year'], df1['carbon_emissions_tg'], marker='o', linewidth=2, markersize=6, 
        color='#1f77b4', label='Carbon Emissions (Ecosystem)')

# Calculate trend line and statistics for ecosystem
z1 = np.polyfit(df1['year'], df1['carbon_emissions_tg'], 1)
p1 = np.poly1d(z1)
trend_line1 = p1(df1['year'])
ax.plot(df1['year'], trend_line1, '--', linewidth=2, color='#1f77b4', alpha=0.7)

# Add average line for ecosystem
avg1 = df1['carbon_emissions_tg'].mean()
ax.axhline(y=avg1, color='#1f77b4', linestyle=':', linewidth=1.5, alpha=0.4)

# Plot 2: Carbon (by gas)
ax.plot(df2['year'], df2['carbon_c_tg'], marker='s', linewidth=2, markersize=6,
        color='#ff7f0e', label='Carbon Emissions (Monthly)')

# Calculate trend line and statistics for C by gas
z2 = np.polyfit(df2['year'], df2['carbon_c_tg'], 1)
p2 = np.poly1d(z2)
trend_line2 = p2(df2['year'])
ax.plot(df2['year'], trend_line2, '--', linewidth=2, color='#ff7f0e', alpha=0.7)

# Add average line for C by gas
avg2 = df2['carbon_c_tg'].mean()
ax.axhline(y=avg2, color='#ff7f0e', linestyle=':', linewidth=1.5, alpha=0.4)

# Add text with statistics for both series
slope1 = z1[0]
slope2 = z2[0]
text_str = (f'Carbon Emissions (Ecosystem):\n'
            f'  Slope: {slope1:.4f} Tg/year\n'
            f'  Mean: {avg1:.4f} Tg\n\n'
            f'Carbon Emissions (Monthly):\n'
            f'  Slope: {slope2:.4f} Tg/year\n'
            f'  Mean: {avg2:.4f} Tg')
ax.text(0.02, 0.98, text_str, transform=ax.transAxes, fontsize=9,
        verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.85))

ax.set_xlabel('Year', fontsize=11)
ax.set_ylabel('Carbon Emissions (Tg)', fontsize=11)
ax.set_title('Annual Carbon Emissions Comparison: Ecosystem vs Monthly (GFED5.1)', fontsize=12)
ax.grid(True, alpha=0.3)
ax.legend(loc='lower right', fontsize=10)

plt.tight_layout()

plt.savefig(os.path.join('outputs', 'exploratory', 'gfed', 'carbon_emissions_comparison.png'), dpi=300, bbox_inches='tight')
print(f"\nPlot saved to outputs/exploratory/gfed/carbon_emissions_comparison.png")

plt.show()
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read GFED by gas emissions CSV
csv_path = os.path.join('outputs', 'gfed_emissions_in_alaska_by_gas_gram.csv')
df_gas = pd.read_csv(csv_path)

# Filter to 2002-2023
df_gas = df_gas[(df_gas['year'] >= 2002) & (df_gas['year'] <= 2023)]

# Global Warming Potentials (100-year)
gwp_co2 = 1
gwp_co = 1.9
gwp_ch4 = 27

# Convert individual gases to CO2e (g to g CO2e)
# All columns assumed to be in grams
df_gas['CO2_co2e_tg'] = df_gas['CO2'] / 1e12  # grams to Tg
df_gas['CO_co2e_tg'] = (df_gas['CO'] * gwp_co) / 1e12  # grams to Tg with GWP
df_gas['CH4_co2e_tg'] = (df_gas['CH4'] * gwp_ch4) / 1e12  # grams to Tg with GWP
df_gas['C_co2e_tg'] = (df_gas['C'] * 44/12) / 1e12  # grams C to grams CO2e to Tg

# Sum of trace gases (CH4, CO, CO2)
df_gas['trace_sum_co2e_tg'] = df_gas['CO2_co2e_tg'] + df_gas['CO_co2e_tg'] + df_gas['CH4_co2e_tg']

# Create comprehensive plot
fig, axes = plt.subplots(2, 1, figsize=(12, 10))

# ===== Plot 1: Individual trace gases as CO2e =====
ax = axes[0]

# Plot each gas
ax.plot(df_gas['year'], df_gas['CO2_co2e_tg'], marker='o', linewidth=2, markersize=5,
        color='#1f77b4', label='CO₂ (GWP=1)', alpha=0.8)
ax.plot(df_gas['year'], df_gas['CO_co2e_tg'], marker='s', linewidth=2, markersize=5,
        color='#ff7f0e', label='CO (GWP=1.9)', alpha=0.8)
ax.plot(df_gas['year'], df_gas['CH4_co2e_tg'], marker='^', linewidth=2, markersize=5,
        color='#d62728', label='CH₄ (GWP=27)', alpha=0.8)

ax.set_xlabel('Year', fontsize=11)
ax.set_ylabel('Emissions (Tg CO₂e)', fontsize=11)
ax.set_title('Annual Trace Gas Emissions by Gas (GFED5.1 - Alaska)', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(loc='best', fontsize=10)

# ===== Plot 2: Carbon component vs Sum of trace gases =====
ax = axes[1]

# Plot C converted to CO2e
ax.plot(df_gas['year'], df_gas['C_co2e_tg'], marker='o', linewidth=2, markersize=6,
        color='#2ca02c', label='C (as CO₂e: C × 44/12)', alpha=0.8)

# Add trend line for C
z_c = np.polyfit(df_gas['year'], df_gas['C_co2e_tg'], 1)
p_c = np.poly1d(z_c)
trend_c = p_c(df_gas['year'])
ax.plot(df_gas['year'], trend_c, '--', linewidth=2, color='#2ca02c', alpha=0.6)
slope_c = z_c[0]
avg_c = df_gas['C_co2e_tg'].mean()

# Plot sum of trace gases (CH4 + CO + CO2)
ax.plot(df_gas['year'], df_gas['trace_sum_co2e_tg'], marker='s', linewidth=2, markersize=6,
        color='#9467bd', label='Sum: CH₄ + CO + CO₂ (with GWPs)', alpha=0.8)

# Add trend line for sum
z_sum = np.polyfit(df_gas['year'], df_gas['trace_sum_co2e_tg'], 1)
p_sum = np.poly1d(z_sum)
trend_sum = p_sum(df_gas['year'])
ax.plot(df_gas['year'], trend_sum, '--', linewidth=2, color='#9467bd', alpha=0.6)
slope_sum = z_sum[0]
avg_sum = df_gas['trace_sum_co2e_tg'].mean()

# Add long-term mean lines
ax.axhline(y=avg_c, color='#2ca02c', linestyle=':', linewidth=1.5, alpha=0.4)
ax.axhline(y=avg_sum, color='#9467bd', linestyle=':', linewidth=1.5, alpha=0.4)

# Add text with statistics
text_str = (f'C (as CO₂e):\n'
            f'  Slope: {slope_c:.4f} Tg/year\n'
            f'  Mean: {avg_c:.4f} Tg\n\n'
            f'Trace Sum (CH₄+CO+CO₂):\n'
            f'  Slope: {slope_sum:.4f} Tg/year\n'
            f'  Mean: {avg_sum:.4f} Tg')
ax.text(0.02, 0.98, text_str, transform=ax.transAxes, fontsize=10,
        verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.85))

ax.set_xlabel('Year', fontsize=11)
ax.set_ylabel('Emissions (Tg CO₂e)', fontsize=11)
ax.set_title('Carbon Component vs Sum of Trace Gases (GFED5.1 - Alaska)', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(loc='lower right', fontsize=10)

plt.tight_layout()
plt.savefig(os.path.join('outputs', 'exploratory', 'gfed', 'gfed_trace_gases_breakdown.png'), dpi=300, bbox_inches='tight')
plt.show()

print(f"Plot saved to outputs/exploratory/gfed/gfed_trace_gases_breakdown.png")
print(f"\nSummary Statistics (2001-2024):")
print(f"{'='*60}")
print(f"{'Metric':<30} {'Mean':>12} {'Trend':>12}")
print(f"{'='*60}")
print(f"{'CO₂ (Tg CO₂e)':<30} {df_gas['CO2_co2e_tg'].mean():>12.4f} {np.polyfit(df_gas['year'], df_gas['CO2_co2e_tg'], 1)[0]:>12.4f}/yr")
print(f"{'CO (Tg CO₂e)':<30} {df_gas['CO_co2e_tg'].mean():>12.4f} {np.polyfit(df_gas['year'], df_gas['CO_co2e_tg'], 1)[0]:>12.4f}/yr")
print(f"{'CH₄ (Tg CO₂e)':<30} {df_gas['CH4_co2e_tg'].mean():>12.4f} {np.polyfit(df_gas['year'], df_gas['CH4_co2e_tg'], 1)[0]:>12.4f}/yr")

print(f"{'C (Tg CO₂e)':<30} {avg_c:>12.4f} {slope_c:>12.4f}/yr")
print(f"{'='*60}")
print(f"{'Trace Sum (Tg CO₂e)':<30} {avg_sum:>12.4f} {slope_sum:>12.4f}/yr")
```
